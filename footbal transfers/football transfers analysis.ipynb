```python
import pandas as pd

df=pd.read_csv("football_transfers_2021.csv")
df.dropna(inplace = True)
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(df.groupby(['age'])['age'].count())
```

```python
x = df['age']
plt.hist(x)
plt.show() 
```

```python
print(df.groupby(['country_origin'])['country_origin'].count())
```

```python
np_array = df['country_origin'].to_numpy()
np_array
```

```python
# initializing dict to store frequency of each element
elements_count = {}
# iterating over the elements for frequency
for element in np_array:
   # checking whether it is in the dict or not
   if element in elements_count:
      # incerementing the count by 1
      elements_count[element] += 1
   else:
      # setting the count to 1
      elements_count[element] = 1
# printing the elements frequencies
for key, value in elements_count.items():
   print(f"{key}: {value}")
```

```python
Origin_Country = elements_count.keys()
Players = elements_count.values()  

fig=plt.figure(figsize = (20, 10)) 
plt.bar(Origin_Country, Players, color ='magenta',width = 0.4)
plt.xlabel("Origin_Country") 
plt.ylabel("Players") 
plt.title("Players Origin Countries") 
plt.show()
```
